## Fundamento teórico

### Naive Bayes

El clasificador **Naive Bayes** es un modelo probabilístico basado en el **teorema de Bayes** con una fuerte asunción de independencia entre las variables dado el valor de la clase. Su objetivo es calcular la probabilidad posterior de una clase $y$ dado un vector de atributos $\mathbf{x} = (x_1, x_2, \dots, x_n)$:

$$
P(y \mid \mathbf{x}) = \frac{P(y) \cdot P(\mathbf{x} \mid y)}{P(\mathbf{x})}
$$

Dado que $P(\mathbf{x})$ es constante para todas las clases en un problema de clasificación, la predicción se basa en:

$$
\hat{y} = \arg\max_y \; P(y) \prod_{i=1}^{n} P(x_i \mid y)
$$

La clave está en estimar adecuadamente las distribuciones de las verosimilitudes $P(x_i \mid y)$.


### Gaussian Naive Bayes

En el enfoque **Gaussian Naive Bayes (GNB)**, se hace la **asunción paramétrica** de que cada variable numérica sigue una distribución **normal univariada** dentro de cada clase. Es decir:

$$
P(x_i \mid y) = \mathcal{N}(x_i \mid \mu_{iy}, \sigma^2_{iy}) = \frac{1}{\sqrt{2\pi \sigma^2_{iy}}} \exp\left( -\frac{(x_i - \mu_{iy})^2}{2\sigma^2_{iy}} \right)
$$

Donde $\mu_{iy}$ y $\sigma^2_{iy}$ se estiman directamente desde los datos de entrenamiento.

Este supuesto simplifica el modelo, pero puede ser problemático si los datos no siguen una distribución gaussiana, por ejemplo, si son **multimodales, sesgados o tienen colas pesadas**.


### KDE como alternativa para estimar la verosimilitud

Una alternativa interesante a asumir una distribución normal es usar un enfoque **no paramétrico** para estimar la densidad de probabilidad: la **Kernel Density Estimation (KDE)**.

En este método, la verosimilitud $P(x_i \mid y)$ se estima directamente a partir de los datos de entrenamiento, sumando pequeñas funciones kernel centradas en cada observación. La forma general es:

$$
\hat{f}(x) = \frac{1}{n h} \sum_{i=1}^{n} K\left( \frac{x - x_i}{h} \right)
$$

Donde:
- $K$ es la función kernel (por ejemplo, gaussiana, triangular, rectangular),
- $h$ es el parámetro de suavizado o **bandwidth**,
- $x_i$ son las observaciones de la clase $y$.

Este enfoque permite capturar formas de distribución **arbitrarias** y potencialmente más realistas que una gaussiana simple, lo cual podría mejorar la clasificación en contextos donde la suposición de normalidad no se cumple.

En este proyecto, exploraremos esta idea usando tres variantes de KDE y comparándolas contra Gaussian Naive Bayes, evaluando cuál ofrece un mejor desempeño en el contexto de mantenimiento predictivo.

### 2. Evaluación de tres métodos de KDE

Cada grupo deberá comparar los siguientes tres métodos de estimación de densidad para calcular la verosimilitud $P(x_i \mid y)$.



#### a. KDE clásico con kernel gaussiano + optimización de `bandwidth`

- Se utiliza un kernel gaussiano para estimar la densidad:
  
  $$
  \hat{f}(x) = \frac{1}{n h \sqrt{2\pi}} \sum_{i=1}^{n} \exp\left( -\frac{(x - x_i)^2}{2h^2} \right)
  $$

- El parámetro `bandwidth` $h$ controla la suavidad de la estimación.
- Para seleccionar un valor apropiado de $h$, se debe usar un método de optimización como **Grid Search** o preferiblemente **optimización bayesiana**, evaluando el rendimiento del modelo en términos de **AUC ROC** mediante validación cruzada (al menos 5 folds).



#### b. KDE tipo Parzen (ventanas simples)

- En este enfoque, se reemplaza el kernel gaussiano por una ventana de forma simple, como rectangular o triangular.

  - **Ventana rectangular (tophat):**
    $$
    K(u) = \frac{1}{2} \cdot \mathbb{I}(|u| \leq 1)
    $$

  - **Ventana triangular:**
    $$
    K(u) = (1 - |u|) \cdot \mathbb{I}(|u| \leq 1)
    $$

- El modelo general de densidad estimada sigue siendo:

  $$
  \hat{f}(x) = \frac{1}{n h} \sum_{i=1}^{n} K\left( \frac{x - x_i}{h} \right)
  $$

- El ancho de ventana $h$ debe definirse manualmente y mantenerse fijo.
- Puede implementarse directamente o usando `sklearn.neighbors.KernelDensity` con `kernel='tophat'` o `'linear'`.


#### c. KDE con regla de Silverman

- Se utiliza un kernel gaussiano, pero el `bandwidth` se calcula automáticamente con la **regla de Silverman**:

  $$
  h = 1.06 \cdot \hat{\sigma} \cdot n^{-1/5}
  $$

- La estimación de la densidad sigue la misma forma que el KDE gaussiano clásico:

  $$
  \hat{f}(x) = \frac{1}{n h \sqrt{2\pi}} \sum_{i=1}^{n} \exp\left( -\frac{(x - x_i)^2}{2h^2} \right)
  $$

- Este método no requiere ajuste de hiperparámetros.
- Se puede implementar fácilmente con `scipy.stats.gaussian_kde`.

## Preprocesamiento

- Se eliminan identificadores (`UDI`, `Product ID`).
- `Type` se codifica como categoría ordinal para poder modelar su probabilidad condicional.
- Se diferencian variables **continuas** (temperaturas, velocidad, torque, desgaste) de variables **categóricas/boleanas** (tipo y fallas específicas).
- Las variables continuas se escalarán con `StandardScaler` dentro de cada partición de validación para favorecer el ajuste KDE.


```python
def preprocess_dataframe(df: pd.DataFrame):
    df = df.copy()
    df['Type'] = df['Type'].astype('category')
    df['Type_code'] = df['Type'].cat.codes
    drop_cols = ['UDI', 'Product ID', 'Type']
    feature_cols = [c for c in df.columns if c not in drop_cols + [target_col]]
    X = df[feature_cols]
    y = df[target_col].astype(int)
    metadata = {
        'feature_cols': feature_cols,
        'numeric_features': ['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]'],
        'categorical_features': ['Type_code', 'TWF', 'HDF', 'PWF', 'OSF', 'RNF'],
        'type_categories': df['Type'].cat.categories.tolist()
    }
    return X, y, metadata

X, y, metadata = preprocess_dataframe(raw_df)
metadata
```

### Función de Preprocesamiento

**Paso a paso:**
1.  Se define una función `preprocess_dataframe` para encapsular la limpieza de datos.
2.  **Eliminación de IDs:** Se descartan `UDI` y `Product ID` porque son identificadores únicos y no tienen poder predictivo.
3.  **Codificación Categórica:** Se convierte la variable `Type` (que tiene valores 'L', 'M', 'H') en códigos numéricos (`0`, `1`, `2`) usando `.cat.codes`.
4.  **Separación X/y:** Se separa el `DataFrame` en `X` (matriz de características) e `y` (vector objetivo).
5.  **Creación de Metadatos:** Se crea un diccionario `metadata`. Esta es una práctica excelente:
    * Almacena explícitamente qué columnas son `numeric_features` (las que recibirán escalado y KDE) y cuáles son `categorical_features` (las que Naive Bayes tratará como discretas).
    * Esto hace que el resto del *pipeline* (SMOTENC y el clasificador) sea mucho más limpio, ya que solo necesitamos pasar este diccionario.

## Utilidades para escalado, sobremuestreo y validación cruzada

```python
def scale_features(X_train: pd.DataFrame, X_valid: pd.DataFrame, numeric_features):
    scaler = StandardScaler()
    X_train_scaled = X_train.copy()
    X_valid_scaled = X_valid.copy()
    X_train_scaled[numeric_features] = scaler.fit_transform(X_train[numeric_features])
    X_valid_scaled[numeric_features] = scaler.transform(X_valid[numeric_features])
    return X_train_scaled, X_valid_scaled, scaler


def apply_smote(X_train: pd.DataFrame, y_train: pd.Series, categorical_features, random_state=42):
    categorical_indices = [X_train.columns.get_loc(col) for col in categorical_features]
    smote = SMOTENC(categorical_features=categorical_indices, random_state=random_state)
    X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
    X_resampled = pd.DataFrame(X_resampled, columns=X_train.columns)
    for col in categorical_features:
        X_resampled[col] = X_resampled[col].round().astype(int)
    return X_resampled, pd.Series(y_resampled, name=y_train.name)


def cross_validate_model(model, X_data, y_data, numeric_features, categorical_features, cv):
    auc_scores = []
    fit_times = []
    for train_idx, valid_idx in cv.split(X_data, y_data):
        X_tr, X_val = X_data.iloc[train_idx].copy(), X_data.iloc[valid_idx].copy()
        y_tr, y_val = y_data.iloc[train_idx], y_data.iloc[valid_idx]
        X_tr_scaled, X_val_scaled, _ = scale_features(X_tr, X_val, numeric_features)
        X_tr_bal, y_tr_bal = apply_smote(X_tr_scaled, y_tr, categorical_features)
        model_cv = clone(model)
        start = time.time()
        model_cv.fit(X_tr_bal, y_tr_bal)
        fit_times.append(time.time() - start)
        y_pred_proba = model_cv.predict_proba(X_val_scaled)[:, 1]
        auc_scores.append(roc_auc_score(y_val, y_pred_proba))
    return np.array(auc_scores), np.array(fit_times)
```

### Funciones de Utilidad del Pipeline

Se definieron tres funciones auxiliares que forman el núcleo de nuestro protocolo de experimentación.

**1. `scale_features(X_train, X_valid, numeric_features)`**
* **Qué hace:** Aplica `StandardScaler` (estandarización) solo a las columnas numéricas.
* **Por qué:** El KDE es sensible a la escala de las variables. Estandarizar (media=0, std=1) asegura que el `bandwidth` (ancho de banda) tenga un efecto comparable en todas las características.
* **Prevención de *Data Leakage*:** Sigue la práctica correcta: ajusta (`fit_transform`) el *scaler* **solo** en los datos de entrenamiento (`X_train`) y luego usa ese mismo *scaler* para transformar (`transform`) los datos de validación/prueba (`X_valid`).

**2. `apply_smote(X_train, y_train, categorical_features, ...)`**
* **Qué hace:** Aplica `SMOTENC` a los datos de entrenamiento.
* **Por qué:** `SMOTENC` crea muestras sintéticas de la clase minoritaria (fallas) para balancear el *dataset*. Es la técnica correcta aquí porque sabe que debe interpolar entre las características numéricas, pero debe seleccionar (no interpolar) valores de las características categóricas.
* **Ubicación:** Esto se aplica **después** del escalado y **solo** al set de entrenamiento, **dentro** de cada *fold* de validación cruzada.

**3. `cross_validate_model(model, X_data, y_data, ...)`**
* **Qué hace:** Es el motor de la validación cruzada (5 *folds*).
* **Protocolo por *Fold*:** Para cada una de las 5 particiones:
    1.  Divide los datos de entrenamiento (`X_data`) en `X_tr` (entrenamiento del *fold*) y `X_val` (validación del *fold*).
    2.  Escala `X_tr` y `X_val` (usando `scale_features`).
    3.  Aplica `SMOTENC` **solo** a `X_tr` y `y_tr` (obteniendo `X_tr_bal`, `y_tr_bal`).
    4.  Entrena el modelo (`model.fit`) usando los datos balanceados (`X_tr_bal`, `y_tr_bal`).
    5.  Evalúa el modelo (calcula `roc_auc_score`) usando `X_val_scaled` (los datos de validación originales, **sin** balancear) y `y_val`.
* **Resultado:** Devuelve un arreglo de 5 puntajes de AUC y 5 tiempos de entrenamiento, permitiéndonos calcular la media y la desviación estándar del rendimiento.

## Implementación: Naive Bayes con KDE

```python
class KDENaiveBayes(BaseEstimator, ClassifierMixin):
    """Clasificador Naive Bayes con estimación de densidad Kernel por característica continua."""

    def __init__(
        self,
        continuous_features,
        categorical_features=None,
        kernel='gaussian',
        bandwidth=0.2,
        bandwidth_strategy='fixed',
        categorical_smoothing=1.0,
        min_bandwidth=1e-3,
        silverman_scale=1.0,
    ):
        self.continuous_features = continuous_features
        self.categorical_features = categorical_features
        self.kernel = kernel
        self.bandwidth = bandwidth
        self.bandwidth_strategy = bandwidth_strategy
        self.categorical_smoothing = categorical_smoothing
        self.min_bandwidth = min_bandwidth
        self.silverman_scale = silverman_scale

    def _prepare_dataframe(self, X):
        if isinstance(X, pd.DataFrame):
            return X.copy()
        return pd.DataFrame(X, columns=self.feature_names_)

    def _compute_bandwidth(self, values):
        if self.bandwidth_strategy == 'silverman':
            n = len(values)
            if n <= 1:
                return self.min_bandwidth
            std = np.std(values, ddof=1)
            scale = min(std, iqr(values) / 1.34) if std > 0 else iqr(values) / 1.34
            if not np.isfinite(scale) or scale <= 0:
                scale = std if std > 0 else self.min_bandwidth
            h = 0.9 * scale * (n ** (-1 / 5))
            h = max(self.min_bandwidth, h)
            return h * self.silverman_scale
        return max(self.min_bandwidth, self.bandwidth)

    def fit(self, X, y):
        X_df = self._prepare_dataframe(X)
        y = np.asarray(y)
        self.feature_names_ = list(X_df.columns)
        self.feature_to_index_ = {name: idx for idx, name in enumerate(self.feature_names_)}
        self.classes_, y_indices = np.unique(y, return_inverse=True)
        self.class_prior_ = np.array([
            np.mean(y == cls) for cls in self.classes_
        ])
        self.continuous_features_ = list(self.continuous_features)
        self.categorical_features_ = list(self.categorical_features) if self.categorical_features is not None else []
        self.cont_indices_ = [self.feature_to_index_[name] for name in self.continuous_features_]
        self.cat_indices_ = [self.feature_to_index_[name] for name in self.categorical_features_]

        self.kdes_ = {}
        self.kde_bandwidths_ = {}
        self.categorical_log_probs_ = {}
        self.categorical_values_ = {
            idx: np.unique(X_df.iloc[:, idx]) for idx in self.cat_indices_
        }

        for cls_idx, cls in enumerate(self.classes_):
            cls_mask = y == cls
            X_cls = X_df.iloc[cls_mask, :].to_numpy()
            kdes_cls = []
            bw_cls = []
            for f_idx in self.cont_indices_:
                values = X_cls[:, f_idx].reshape(-1, 1)
                bw = self._compute_bandwidth(X_cls[:, f_idx])
                kde = KernelDensity(kernel=self.kernel, bandwidth=bw)
                kde.fit(values)
                kdes_cls.append(kde)
                bw_cls.append(bw)
            self.kdes_[cls] = kdes_cls
            self.kde_bandwidths_[cls] = bw_cls

            cat_log_probs = {}
            for cat_idx in self.cat_indices_:
                values = X_cls[:, cat_idx]
                unique_all = self.categorical_values_[cat_idx]
                counts = pd.Series(values).value_counts().reindex(unique_all, fill_value=0).astype(float)
                alpha = self.categorical_smoothing
                probs = (counts + alpha) / (counts.sum() + alpha * len(unique_all))
                log_probs = np.log(probs)
                fallback = math.log(alpha / (counts.sum() + alpha * len(unique_all)))
                cat_log_probs[cat_idx] = {
                    'log_probs': log_probs.to_dict(),
                    'fallback': fallback,
                }
            self.categorical_log_probs_[cls] = cat_log_probs
        return self

    def _compute_log_likelihood(self, X_df):
        n_samples = X_df.shape[0]
        log_likelihood = np.zeros((n_samples, len(self.classes_)))
        X_values = X_df.to_numpy()
        log_prob_floor = -1000.0
        for cls_pos, cls in enumerate(self.classes_):
            log_prob = np.log(self.class_prior_[cls_pos])
            for kde, f_idx in zip(self.kdes_[cls], self.cont_indices_):
                log_densities = kde.score_samples(X_values[:, f_idx].reshape(-1, 1))
                log_densities[np.isneginf(log_densities)] = log_prob_floor
                log_prob += log_densities
            for cat_idx in self.cat_indices_:
                cat_info = self.categorical_log_probs_[cls][cat_idx]
                value_map = cat_info['log_probs']
                fallback = cat_info['fallback']
                vals = X_values[:, cat_idx]
                log_prob += np.array([value_map.get(val, fallback) for val in vals])
            log_likelihood[:, cls_pos] = log_prob
        return log_likelihood

    def predict_proba(self, X):
        check_is_fitted(self, 'classes_')
        X_df = self._prepare_dataframe(X)
        log_likelihood = self._compute_log_likelihood(X_df)
        log_norm = log_likelihood - logsumexp(log_likelihood, axis=1, keepdims=True)
        return np.exp(log_norm)

    def predict(self, X):
        proba = self.predict_proba(X)
        return self.classes_[np.argmax(proba, axis=1)]

    def score(self, X, y):
        return np.mean(self.predict(X) == y)
```

### Análisis: Implementación del Clasificador Naive Bayes con KDE

Esta es la pieza central del proyecto. Se crea un clasificador personalizado que sigue la API de `sklearn` (`BaseEstimator`, `ClassifierMixin`).

**Componentes Clave:**

* **`__init__(self, ...)`:**
    * El constructor almacena los hiperparámetros:
        * `continuous_features` y `categorical_features`: Listas de columnas.
        * `kernel`: El tipo de kernel para `sklearn.KernelDensity` (ej. 'gaussian', 'tophat').
        * `bandwidth`: El valor de `h` (ancho de banda).
        * `bandwidth_strategy`: Permite cambiar entre `h` 'fixed' (fijo) o 'silverman' (calculado automáticamente).

* **`_compute_bandwidth(self, values)`:**
    * Implementa la **Regla de Silverman** (`strategy='silverman'`).
    * Es una heurística estándar para elegir un `bandwidth` "razonable" automáticamente, basado en la desviación estándar (o el rango intercuartílico, para robustez) y el número de muestras.
    * Es la alternativa a la costosa búsqueda por *Grid Search*.

* **`fit(self, X, y)`:**
    * Aquí ocurre el "entrenamiento" del modelo Naive Bayes.
    1.  **Calcula Priors `P(y)`:** Calcula la probabilidad *a priori* de cada clase (ej. `P(Falla) = 0.5` en el set balanceado).
    2.  **Entrena Verosimilitudes Categóricas `P(x_i | y)`:** Para cada clase `y` y cada *feature* categórica `x_i`, calcula la tabla de probabilidades. (ej. `P(Type='L' | Falla)`). Usa suavizado de Laplace (`categorical_smoothing`) para evitar probabilidades de cero si una categoría no aparece.
    3.  **Entrena Verosimilitudes Continuas `P(x_i | y)`:** Para cada clase `y` y cada *feature* continua `x_i`:
        * Calcula el `bandwidth` (ya sea fijo o usando Silverman).
        * Crea una instancia de `sklearn.KernelDensity` con ese kernel y `bandwidth`.
        * Ajusta (`.fit()`) el KDE solo con los datos de esa *feature* y esa clase.
    * Almacena todos los KDEs y tablas de probabilidad en diccionarios (`self.kdes_`, `self.categorical_log_probs_`).

* **`_compute_log_likelihood(self, X_df)`:**
    * Calcula la "puntuación" de Naive Bayes para cada muestra `X` y cada clase `y`.
    * Funciona en **espacio logarítmico** para estabilidad numérica.
    * Para una muestra dada, la puntuación de una clase es:
        `log(P(y)) + Σ log(P(x_cat | y)) + Σ log(P(x_cont | y))`
    * El `log(P(x_cont | y))` se obtiene directamente del `kde.score_samples()`.
    * **Corrección de NaN (¡Importante!):** Se aplica un *floor* (`LOG_PROB_FLOOR = -1000.0`). Cuando un kernel de soporte finito (como 'tophat') ve un punto fuera de su ventana, `score_samples` devuelve `-inf` (log(0)). Si esto pasa para todas las clases, `logsumexp` calculará `inf - inf = NaN`. Al reemplazar `-inf` por un número muy negativo, evitamos el `NaN` y mantenemos la estabilidad.

* **`predict_proba(self, X)`:**
    * Toma los *log-likelihoods* conjuntos del paso anterior.
    * Usa `logsumexp` para normalizarlos y convertirlos en probabilidades (entre 0 y 1).

## Validación cruzada: referencia Gaussian Naive Bayes

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

baseline_model = GaussianNB()
baseline_aucs, baseline_times = cross_validate_model(
    baseline_model,
    X_train,
    y_train,
    metadata['numeric_features'],
    metadata['categorical_features'],
    cv
)
print(f"AUC promedio: {baseline_aucs.mean():.4f} ± {baseline_aucs.std():.4f}")
print(f"Tiempo medio de entrenamiento: {baseline_times.mean():.3f} s")
```

### Baseline (Gaussian Naive Bayes)

**Paso a paso:**
1.  Se definió un `GaussianNB` estándar de `sklearn` como nuestro modelo de referencia.
2.  Se utilizó la función `cross_validate_model`, que aplica una validación cruzada estratificada de 5 *folds* (particiones).
3.  **Importante:** Dentro de cada *fold*, los datos de entrenamiento se escalaron (`StandardScaler`) y luego se balancearon con `SMOTENC`. El *fold* de validación solo se escaló (usando el *scaler* ajustado en el *fold* de entrenamiento).
4.  Se calculó el **AUC ROC** y el tiempo de entrenamiento para cada *fold*.

**Análisis de Resultados:**
* El `GaussianNB` estándar obtiene un AUC promedio de **0.9871 ± 0.0120**.
* Este valor es nuestra **línea base (baseline)**. Cualquier modelo KDE personalizado que desarrollemos debe superar este puntaje para justificar su complejidad adicional.
* El tiempo de entrenamiento es extremadamente bajo (**0.003 s** por *fold*), lo cual es característico de GNB, ya que solo necesita calcular la media y la varianza de cada característica por clase.

## Búsqueda de bandwidth óptimo para KDE gaussiano

```python
bandwidth_grid = np.array([0.05, 0.1, 0.2, 0.4, 0.8])
results_gaussian = []
for bw in bandwidth_grid:
    model = KDENaiveBayes(
        continuous_features=metadata['numeric_features'],
        categorical_features=metadata['categorical_features'],
        kernel='gaussian',
        bandwidth=bw,
        bandwidth_strategy='fixed',
        categorical_smoothing=1.0,
        min_bandwidth=1e-3,
    )
    aucs, times_model = cross_validate_model(
        model,
        X_train,
        y_train,
        metadata['numeric_features'],
        metadata['categorical_features'],
        cv
    )
    results_gaussian.append({'bandwidth': bw, 'mean_auc': aucs.mean(), 'std_auc': aucs.std(), 'mean_time': times_model.mean()})

gaussian_search_df = pd.DataFrame(results_gaussian)
gaussian_search_df.sort_values(by='mean_auc', ascending=False).head()
```

### Búsqueda de Hiperparámetros (KDE Gaussiano)

**Paso a paso:**
1.  Se definió una grilla de posibles valores para el hiperparámetro `bandwidth` (ancho de banda): `[0.05, 0.1, 0.2, 0.4, 0.8]`.
2.  Para cada valor de `bandwidth`, se ejecutó la misma validación cruzada de 5 *folds* (con escalado y `SMOTENC`) que se usó para el *baseline*, usando nuestro `KDENaiveBayes` con `kernel='gaussian'`.
3.  Se registraron el AUC promedio y el tiempo de entrenamiento para cada `bandwidth`.

**Análisis de Resultados:**
* Los resultados de la tabla (ordenados por `mean_auc`) muestran el impacto directo del `bandwidth` en el rendimiento.
* El mejor rendimiento se obtiene con un `bandwidth` de **0.80**, logrando un AUC de **0.988344**.
* Se observa que todos los anchos de banda probados dan un rendimiento muy similar y ligeramente superior al *baseline* de `GaussianNB` (0.9871). El `bandwidth` más pequeño (0.05) rinde peor, sugiriendo un leve sobreajuste.
* El tiempo de entrenamiento (aprox. 0.02s) es significativamente mayor que el de GNB (0.003s), como era de esperar.

**Análisis de Resultados:**
* **Rendimiento (AUC):** Los modelos KDE con *bandwidth* optimizado (Gaussiano y Parzen, ambos con h=0.800) **superan marginalmente** al *baseline* `GaussianNB` (0.9883 vs 0.9871).
* El **KDE Silverman** (método automático) queda en **último lugar**, rindiendo peor que el *baseline*.
* **Conclusión de CV:** La optimización del `bandwidth` parece aportar una ganancia mínima pero existente. El costo computacional es notablemente mayor (0.022s vs 0.003s). Hasta ahora, los KDE optimizados son los "ganadores" por un estrecho margen.

## Entrenamiento final y evaluación en conjunto de prueba

```python
# Preparación: escalar datos completos de entrenamiento y aplicar SMOTENC
X_train_scaled, X_test_scaled, scaler = scale_features(X_train, X_test, metadata['numeric_features'])
X_train_bal, y_train_bal = apply_smote(X_train_scaled, y_train, metadata['categorical_features'])

models_final = {
    'GaussianNB': GaussianNB(),
    'KDE Gaussiano': KDENaiveBayes(
        continuous_features=metadata['numeric_features'],
        categorical_features=metadata['categorical_features'],
        kernel='gaussian',
        bandwidth=best_gaussian_bw,
        bandwidth_strategy='fixed'
    ),
    'KDE Parzen': KDENaiveBayes(
        continuous_features=metadata['numeric_features'],
        categorical_features=metadata['categorical_features'],
        kernel='tophat',
        bandwidth=best_parzen_bw,
        bandwidth_strategy='fixed'
    ),
    'KDE Silverman': KDENaiveBayes(
        continuous_features=metadata['numeric_features'],
        categorical_features=metadata['categorical_features'],
        kernel='gaussian',
        bandwidth_strategy='silverman'
    )
}

test_results = []
probabilities = {}
for name, model in models_final.items():
    start = time.time()
    model.fit(X_train_bal, y_train_bal)
    fit_time = time.time() - start
    y_proba = model.predict_proba(X_test_scaled)[:, 1]
    auc = roc_auc_score(y_test, y_proba)
    probabilities[name] = y_proba
    test_results.append({'Modelo': name, 'AUC test': auc, 'Tiempo entrenamiento (s)': fit_time})

test_results_df = pd.DataFrame(test_results).sort_values('AUC test', ascending=False)
test_results_df
```

### Evaluación Final en Conjunto de Prueba (Test Set)

**Paso a paso:**
1.  Se tomaron los 4 modelos finalistas (GNB y las 3 variantes de KDE).
2.  Se re-entrenaron una última vez usando **todo** el conjunto de entrenamiento (`X_train` completo), aplicando primero el escalado y luego el balanceo `SMOTENC`.
3.  Se usaron estos modelos para predecir las probabilidades en el conjunto de prueba (`X_test_scaled`), que se mantuvo oculto hasta ahora.
4.  Se calculó el AUC ROC final en estos datos "nunca antes vistos".

**Análisis de Resultados:**
* **¡Resultado Sorprendente!** La tabla de `AUC test` **revierte** los hallazgos de la validación cruzada.
* El **`GaussianNB` (baseline) es el ganador** en el conjunto de prueba, con un AUC de **0.990683**.
* Todos los modelos KDE, que habían ganado marginalmente en la CV, ahora obtienen un rendimiento **inferior** en el *test set* (ej. KDE Gaussiano con 0.986687).
* **Conclusión Clave:** Esto sugiere que la complejidad adicional de los modelos KDE (incluyendo la optimización del `bandwidth`) llevó a un **leve sobreajuste** al conjunto de entrenamiento/validación. El modelo `GaussianNB`, más simple, resultó ser más robusto y **generalizó mejor** a los datos nuevos.

### Densidades estimadas para `Torque [Nm]`

Se ilustra cómo cambia la densidad condicional $P(	ext{Torque} \mid y)$ para la clase positiva bajo cada variante KDE.

```python
feature = 'Torque [Nm]'
feature_idx = metadata['numeric_features'].index(feature)
values = np.linspace(X_test_scaled[feature].min()-1, X_test_scaled[feature].max()+1, 200)

fig, ax = plt.subplots(figsize=(8, 5))
for name, model in models_final.items():
    if not hasattr(model, 'kdes_'):
        continue
    kde_pos = model.kdes_[1][feature_idx]
    log_dens = kde_pos.score_samples(values.reshape(-1, 1))
    ax.plot(values, np.exp(log_dens), label=name)
ax.set_xlabel('Torque [Nm] (escalado)')
ax.set_ylabel('Densidad estimada')
ax.set_title('Comparación de densidades para la clase positiva')
ax.legend()
plt.show()
```

**Análisis de Resultados:**

    * **`KDE Silverman` (Roja):** Es la curva más "nerviosa" y compleja. Claramente se **sobreajustó**. El modelo `GaussianNB`, al ser más simple y "rígido" (asumiendo una campana de Gauss), **ignoró este ruido** y, por lo tanto, fue el que mejor **generalizó** a los datos de prueba.

### 🏆 Conclusión 1: El modelo más simple (`GaussianNB`) fue el ganador

El resultado más sorprendente e importante es que el modelo de referencia, **`GaussianNB`**, fue el que obtuvo el **mejor rendimiento en el conjunto de prueba** (Test Set), con un **AUC de 0.9907**.

Aunque los modelos `KDE` optimizados (Gaussiano y Parzen) ganaron *marginalmente* durante la validación cruzada (CV), no lograron generalizar tan bien. Esto refuta la hipótesis inicial del proyecto: para este *dataset* y con este preprocesamiento, la suposición gaussiana simple fue más robusta.

---

### 📉 Conclusión 2: La flexibilidad del KDE llevó a un sobreajuste (Overfitting)

La flexibilidad que se buscaba con `KDE` resultó ser una desventaja. Los modelos `KDE` (especialmente Silverman) se **sobreajustaron** al ruido del conjunto de entrenamiento.

* **Evidencia:** El gráfico de "Comparación de densidades" muestra esto perfectamente. La curva roja (`KDE Silverman`) es errática y aprendió múltiples picos (como el pico anómalo cerca de -3.0). Esto es un signo claro de sobreajuste al ruido, probablemente amplificado por las muestras sintéticas de `SMOTENC`.
* **Consecuencia:** Este sobreajuste llevó a los modelos `KDE` a tener una **precisión (precision) muy baja** (0.52, 0.36 y un desastroso 0.29 para Silverman), lo que significa que generaron una enorme cantidad de falsas alarmas.

---

### 📈 Conclusión 3: Los modelos KDE se obsesionaron con el Recall

Todos los modelos `KDE` lograron un **`recall` casi perfecto** para la clase positiva (falla), identificando correctamente 66 de las 68 fallas (97.06%).

El problema es que lograron esto volviéndose "demasiado sensibles". Al aprender formas de densidad tan complejas, cualquier pequeña desviación en un sensor se interpretaba como una falla, resultando en la baja precisión que vimos. El `GaussianNB`, al ser más simple, encontró un balance superior entre detectar fallas (`recall`) y no generar falsas alarmas (`precision`), lo que se tradujo en un mejor **AUC** general.

---

### 🤔 Conclusión 4: La optimización del `bandwidth` es crítica y "automático" no es mejor

La búsqueda de hiperparámetros (`Grid Search`) demostró ser efectiva. El `bandwidth` óptimo (h=0.8) para los kernels Gaussiano y Parzen dio un rendimiento superior en CV (0.9883) al del `GaussianNB` (0.9871).

Sin embargo, el método automático (`KDE Silverman`) tuvo el **peor rendimiento de todos** (AUC en CV de 0.9856). Esto demuestra que las heurísticas automáticas como la regla de Silverman no son una bala de plata y pueden fallar, especialmente en datos complejos o sintéticos (post-SMOTENC).

---

### 💡 Conclusión Final del Proyecto

La hipótesis de que un modelo de densidad no paramétrico (`KDE`) superaría a un modelo paramétrico (`GaussianNB`) fue **refutada** por los resultados del conjunto de prueba.

La rigidez del `GaussianNB` fue su mayor fortaleza: **ignoró el ruido** al que los modelos `KDE` se sobreajustaron. En este escenario, la complejidad adicional no aportó valor y, de hecho, perjudicó la capacidad de generalización del modelo. El `GaussianNB` (el modelo más simple y rápido) fue el campeón indiscutible.
